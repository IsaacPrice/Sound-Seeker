```python
import pandas as pd
data = pd.read_csv("AlbumRankings.csv")
data.head()
```

1. Take out the unneeded columns
2. Take out the first row
3. Seperate the columns into the pairs with the name and rating
4. Clean each new dataframe
5. Combine and write to a new csv

```python
# Take out the unneeded columns
keep = ['Midnights', 'Unnamed: 1', 'Utopia', 'Unnamed: 4', 'Pink Tape', 'Unnamed: 7', 'Her Loss', 'Unnamed: 10', '808s & Heartbreak', 'Unnamed: 13']
data = data[keep]
data
```

```python
# Take out the first row
data = data.drop(0, axis='index')
data.head()
```

```python
# Seperate the columns into the pairs with the name and rating
col1 = data[['Midnights', 'Unnamed: 1']]
col2 = data[['Utopia', 'Unnamed: 4']]
col3 = data[['Pink Tape', 'Unnamed: 7']]
col4 = data[['Her Loss', 'Unnamed: 10']]
col5 = data[['808s & Heartbreak', 'Unnamed: 13']]
```

```python
# Clean each new dataframe
col1.dropna(axis='index', inplace=True)
col2.dropna(axis='index', inplace=True)
col3.dropna(axis='index', inplace=True)
col4.dropna(axis='index', inplace=True)
col5.dropna(axis='index', inplace=True)
col1 = col1.rename(columns={'Midnights': 'song', 'Unnamed: 1' : 'value'})
col2 = col2.rename(columns={'Utopia': 'song', 'Unnamed: 4' : 'value'})
col3 = col3.rename(columns={'Pink Tape': 'song', 'Unnamed: 7' : 'value'})
col4 = col4.rename(columns={'Her Loss': 'song', 'Unnamed: 10' : 'value'})
col5 = col5.rename(columns={'808s & Heartbreak': 'song', 'Unnamed: 13' : 'value'})
```

```python
# Combine and write to a csv
new_data = pd.concat([col1, col2, col3, col4, col5], axis='index', join='inner')
new_data.head()
```

```python
# Write to a csv
new_data.to_csv("Cleaned.csv")
```
